```python
from functools import partial

import numpy as np
import seaborn as sns
from ipywidgets import FloatSlider, IntSlider, interact, interactive
from matplotlib import pyplot as plt

import filters
import noise
from utils import imshow, panels
```

```python
WindowSlider = partial(
    IntSlider, value=3, min=1, max=21, step=2, description="Окно"
)
RadiusSlider = partial(FloatSlider, min=0.1, value=1, max=5)
AmountSlider = partial(FloatSlider, value=1, min=0, max=5)
```

```python
sns.set_theme()
```

```python
image = plt.imread('Lenna.png')
# image = plt.imread('lenna.jpg') / 256
```

## Шум Перца и Нормальный

```python
@interact
def inner(
    salt=FloatSlider(value=0.1, min=0, max=0.5, ),
    scale=FloatSlider(value=0.25, min=0, max=1),
    ):
    global salted, gauesd
    left, right = panels()
    imshow(salted := noise.salt(image, salt), left)
    imshow(gauesd := noise.normal(image, scale), right)
```

```python
noised = salted
```

## Фильтр Прямоугольника

```python
@interact
def inner(window=WindowSlider(), radius=RadiusSlider(), amount=AmountSlider()):
    left, right = panels()
    imshow(filtred := filters.average(noised, window), left)
    imshow(filters.unsharp(filtred, radius, amount), right)
```

## Фильтр Сигма

```python
@interact
def inner(
    window=WindowSlider(),
    sigma=FloatSlider(value=0.5, min=0, max=1),
    radius=RadiusSlider(), amount=AmountSlider()
):
    left, right = panels()
    imshow(filtred := filters.soft(noised, window, sigma), left)
    imshow(filters.unsharp(filtred, radius, amount), right)
```

## Фильтр Медианны

```python
@interact
def inner(window=WindowSlider(), radius=RadiusSlider(), amount=AmountSlider()):
    left, right = panels()
    imshow(filtred := filters.median(noised, window), left)
    imshow(filters.unsharp(filtred, radius, amount), right)
```

## Гаус фильтр

```python
@interact
def inner(sigma = FloatSlider(value=1/3, min=0.1, max=20),
          radius=RadiusSlider(), amount=AmountSlider()):
    left, right = panels()
    imshow(filtred := filters.blur(noised, sigma), left)
    imshow(filters.unsharp(filtred, radius, amount), right)
```
